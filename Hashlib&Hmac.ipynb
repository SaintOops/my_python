# Hashlib


```
import hashlib
```

Для хэширования строк предназначен модуль **hashlib**. 

Что такое хэширование?

Хэширование означает ввод информации любой длины и размера в исходной строке и выдачу результата фиксированной длины заданной алгоритмом функции хэширования.


Конструкторы для алгоритмов хэширования, которые всегда присутствуют в этом модуле являются(атрибут **.algorithms_guaranteed**):

 (Биткоин использует SHA-256)

```
hashlib.algorithms_guaranteed
```

```
pass_word = hashlib.sha256(b"password")
pass_word
```

Метод **.new()** позволяет использовать старонние алгоритмы хэширования, например, **ripemd160** .

```
hashlib.new('sha224')
```

```
hashlib.new('sha256')
```

**.name** - каноническое имя хеш-функции

```
pass_word.name
```

**.digest_size** - размер результирующего хэша в байтах.

```
pass_word.digest_size
```

**.block_size** - размер внутреннего блока хеш-алгоритма в байтах.

```
pass_word.block_size
```

**.update( данные )** - обновите хэш-объект байтовым объектом.


Повторные вызовы эквивалентны одиночному вызову с конкатенацией всех аргументов:\

**pass_word.update(b'a'); pass_word.update(b'b')**  ==   **pass_word.update(b('a'+'b'))**

хэшируем строку 'passwordab'

```
pass_word.update(b'a')
pass_word.update(b'b')
```

методы **.digest()** и **.hexdigest()**

digest - выходной результат хэш-функции (шифр)

**.hexdigest()** - зашифрованная строка

**.digest()** - зашифрованная строка с битами безобасности


```
pass_word.hexdigest()
```

```
pass_word.digest()
```

метод **.copy()** для копирования объекта

# Hmac

Модуль реализует алгоритм HMAC

```
import hmac
```

.new( ключ , сообщение , digestmod = '') 

.digest()\
.update( сообщение )\

.copy( )\
.digest_size\
.block_size\

```
import binascii


key = "E49756B4C8FAB4E48222A3E7F3B97CC3"
byte_key = binascii.unhexlify(key)

message = "TEST STRING"
message = message.encode()

obj = hmac.new(byte_key, message, hashlib.sha256)
obj
```

```
obj.digest()
```

**hashlib** содержит в себе метод для получения ключа на основе **hmac**

**.pbkdf2_hmac( хэш-функция, сообщение, ключ, итерации)**

ключ - байт строка

```
dk = hashlib.pbkdf2_hmac('sha256', b'password', b'key', 100000)
dk
```
