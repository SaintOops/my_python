```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/%D1%83%D0%BF%D1%80%D0%B0%D0%B6%D0%BD%D0%B5%D0%BD%D0%B8%D1%8F/data/synscan.csv")
```

```python
df.Info.str.find("[TCP Retransmission]")
df["tcp_retransmission"] = df.Info.str.find("[TCP Retransmission]")  
df["tcp_retransmission"] = np.where(df["tcp_retransmission"] == 0, True, False)
df["Info"] = df["Info"].str.replace(r"[TCP Retransmission]", "", regex=False)
df["Info"] = df["Info"].str.strip()
```

```python
import re
(sequence, mss, index_mss, win, Len, ack, index_ack, start, afterMore) = ([], [], [], [], [], [], [], [], [])
for i in range(2011):
    a = df.Info[i].split()
    start.append(int(a[0]))
    afterMore.append(int(a[2]))
    for j in a[3::]:
        if re.search('\d+', j) is not None:
            if j[:3] == 'Seq':
                sequence.append(int(j[4::]))
            if j[:3] == 'Win':
                win.append(int(j[4::])) 
            if j[:3] == 'Len':
                Len.append(int(j[4::]))
            if j[:3] == 'MSS':
                index_mss.append(i)
                mss.append(int(j[4::]))
            if j[:3] == 'Ack':
                ack.append(int(j[4::]))
                index_ack.append(i)
                
ack1, mss1 = [], []
for i in range(2011):
    if i in index_ack:
        ack1.append(i)
    else:
        ack1.append('')
    if i in index_mss:
        mss1.append(i)
    else:
        mss1.append('')
        
del df['Info']
```

```python
df['seq'] = sequence
df['ack'] = ack1
df['Len'] = Len
df['win'] = win
df['mss'] = mss1
df['start'] = start
df['afterMore'] = afterMore
```
